```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("merged_top500_clean.csv")

# Count occurrences of each publication date
date_counts = df["Publication Date"].value_counts().sort_index()

# Identify dates with incorrect counts
incorrect_counts = date_counts[date_counts != 500]

# Visualization
plt.figure(figsize=(12, 6))
colors = ['green' if count == 500 else 'red' for count in date_counts]

date_counts.plot(kind='bar', color=colors, width=0.8)
plt.axhline(y=500, color='black', linestyle='dashed', linewidth=1)
plt.xlabel("Publication Date")
plt.ylabel("Number of Entries")
plt.title("Verification of Publication Date Counts")
plt.xticks(rotation=90)
plt.show()

# Plot the counts of the publication dates with a shorter graphic
plt.figure(figsize=(8, 4))  # Adjusted height to make the plot shorter

# Plot the bar graph showing Publication Date counts
date_counts = df['Publication Date'].value_counts().sort_index()
date_counts.plot(kind='bar', color='skyblue')

# Add titles and labels
plt.title('Publication Date Counts (Verification)')
plt.xlabel('Publication Date')
plt.ylabel('Count of Entries')
plt.xticks(rotation=45)
plt.tight_layout()  # Ensures everything fits nicely in the figure
plt.show()

# Plot the counts of the publication dates with a shorter graphic while preserving width
plt.figure(figsize=(12, 4))  # Adjusted height to make the plot shorter, width remains larger for readability

# Plot the bar graph showing Publication Date counts
date_counts = df['Publication Date'].value_counts().sort_index()
date_counts.plot(kind='bar', color='green')

# Add titles and labels
plt.title('Publication Date Counts (Verification)')
plt.xlabel('Publication Date')
plt.ylabel('Count of Entries')
plt.xticks(rotation=45)
plt.tight_layout()  # Ensures everything fits nicely in the figure
plt.show()



# Print anomalies, if any
if not incorrect_counts.empty:
    print("Dates with incorrect counts:")
    print(incorrect_counts)
else:
    print("All publication dates have exactly 500 entries.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("merged_top500_clean.csv")

# Convert "Publication Date" column to datetime format
df["Publication Date"] = pd.to_datetime(df["Publication Date"], format="%m/%Y", errors="coerce")

# Check if dates are in the correct order
df["Is_Misordered"] = df["Publication Date"] < df["Publication Date"].shift(1)

# Find rows with misordered dates
misordered_dates = df[df["Is_Misordered"]]

# Print misordered dates (if any)
if not misordered_dates.empty:
    print("Misordered dates found:")
    print(misordered_dates[["Publication Date"]])
else:
    print("No misordered dates found. The dataset is correctly ordered.")

# Visualization
plt.figure(figsize=(12, 5))
plt.plot(df.index, df["Publication Date"], marker="o", linestyle="-", label="Publication Dates", color="blue")
plt.scatter(misordered_dates.index, misordered_dates["Publication Date"], color="red", label="Misordered Dates", zorder=3)

plt.xlabel("Row Index")
plt.ylabel("Publication Date")
plt.title("Publication Date Order Check")
plt.legend()
plt.grid()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv("merged_top500_clean.csv")

# Convert "Publication Date" to datetime format
df["Publication Date"] = pd.to_datetime(df["Publication Date"], format="%m/%Y", errors="coerce")

# Filter data for the two selected publication dates (June 1993 and November 2024)
df_jun93 = df[df["Publication Date"] == "1993-06-01"]
df_nov24 = df[df["Publication Date"] == "2024-11-01"]

# Create scatter plots with individual y-axes and log scale for x-axis
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# June 1993 plot (left)
axes[0].scatter(df_jun93["Rank"], df_jun93["RMax (Flop/s)"], color="blue", alpha=0.7)
axes[0].set_title("Rank vs RMax (Flop/s) - June 1993")
axes[0].set_xlabel("Rank (Log Scale)")
axes[0].set_ylabel("RMax (Flop/s)")
axes[0].set_xscale("log")  # Apply logarithmic scale to x-axis
axes[0].grid()

# November 2024 plot (right)
axes[1].scatter(df_nov24["Rank"], df_nov24["RMax (Flop/s)"], color="red", alpha=0.7)
axes[1].set_title("Rank vs RMax (Flop/s) - November 2024")
axes[1].set_xlabel("Rank (Log Scale)")
axes[1].set_ylabel("RMax (Flop/s)")
axes[1].set_ylim([0, df_nov24["RMax (Flop/s)"].max() * 1.1])  # Adjust the y-axis for better visibility
axes[1].set_xscale("log")  # Apply logarithmic scale to x-axis
axes[1].grid()

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming df is your DataFrame

# Define the columns to check for missing data and summation
efficiency_columns = ['Efficiency (%)', 'Effeciency (%)']

# Replace 0s with NaN for the selected efficiency columns (to treat them as missing)
df[efficiency_columns] = df[efficiency_columns].replace(0, np.nan)

# Calculate Total Efficiency as the sum of the two columns, handling NaN values
df['Total Efficiency (%)'] = df['Efficiency (%)'].fillna(0) + df['Effeciency (%)'].fillna(0)

# Calculate completeness for each column (non-missing values / total rows)
columns_to_check = ['RMax (Flop/s)', 'RPeak (Flop/s)', 'Power (W)', 'Total Efficiency (%)']
df_copy = df[columns_to_check].replace(0, np.nan)

# Check missing data for individual columns
missing_data = df_copy.isnull().sum()

# Calculate completeness for each column
completeness = 1 - missing_data / len(df)

# Plotting the completeness of each column as a percentage, but only showing Total Efficiency
plt.figure(figsize=(10, 6))

# Plot all columns but only show Total Efficiency in the graph
completeness[['RMax (Flop/s)', 'RPeak (Flop/s)', 'Power (W)', 'Total Efficiency (%)']].plot(kind='bar', color=['skyblue', 'lightgreen', 'orange', 'purple'])
plt.title('Data Completeness for Selected Columns (0 treated as missing)')
plt.ylabel('Completeness (%)')
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.xlabel('Select Columns')
plt.show()

# Plotting the completeness of each column as a percentage, but only showing Total Efficiency
plt.figure(figsize=(8, 4))  # Adjusted height to make the plot shorter

# Plot all columns but only show Total Efficiency in the graph
completeness[['RMax (Flop/s)', 'RPeak (Flop/s)', 'Power (W)', 'Total Efficiency (%)']].plot(kind='bar', color=['skyblue', 'lightgreen', 'orange', 'purple'])
plt.title('Data Completeness for Selected Columns (0 treated as missing)')
plt.ylabel('Completeness (%)')
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.xlabel('Select Columns')
plt.tight_layout()  # Ensures everything fits nicely in the figure
plt.show()

# Print out the missing data and completeness for reference
print(f"Missing data count for each column after treating 0 as missing (for visualization only):\n{missing_data}")
print(f"\nCompleteness (in %) for each column:\n{completeness * 100}")
```
